Its the process of importing data from different platforms or sources(excel, CSV, SQL databases, etc.) for analysis and reporting.

It is considered the first step of analysis because all reports, dashboards, and visualizations depend on the data being available inside Power BI. Without loading the data, analysis cannot begin.

Q2.

**Load**

- Imports data directly.

-  Skips data cleaning and transformation.

- Directly goes to report view.

- Faster when data is already clean.

**Transform data**

- opened in power query tab.

- Used when data contains errors, duplicates, or null values.

- Used for cleaning and modifying.

- alternate for opening other platforms for cleaning and fixing errors

-  can be used as a final place for cleaning data before shaping and preparing for analysis and report making.

*From this we can understand that load is used when the data is ready for reporting and analysis and transform is used when the data needs a cleaning or final touch up before loading*

**Steps to Identify and Remove Duplicates in Power BI**

**1.** Load data
- open power BI in your desktop

- click home > get data

- select the excel file

- Click Transform Data to open Power Query Editor

**2.** Select Required Columns

Hold Ctrl and select:

- Date

- Country

- State

**3.** Remove Duplicates

- Go to the Home tab

- Click Remove Rows

- Select Remove Duplicates

**4.** Verify the Data

Check:

- Duplicate rows are removed

- Total row count is reduced correctly

- Important records are not deleted accidentally

**5.** Apply Changes

Click:

Close & Apply

The cleaned dataset will now load into Power BI.
